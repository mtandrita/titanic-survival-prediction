# **Titanic Dataset:**

## Step 1: Load the data

```python
import pandas as pd

# Load the data
train = pd.read_csv('titanic_train.csv')  # Training data (contains "Survived" column)
test = pd.read_csv('titanic_test.csv')  # Test data (we will predict "Survived" for this)


```

## Step 2: Explore the data

```python
# Show first few rows
train.head()
```

```python
train.info() # Shows column names and data types

```

```python
train.describe() # Mathematically describes numerical columns
```

```python
train.isnull().sum() # Shows how many missing values in each column
```

 ## Step 3: Clean and preprocess data

```python
# Fill missing age with median
train['Age'].fillna(train['Age'].median(), inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)

# Fill missing embarked with mode
train['Embarked'].fillna(train['Embarked'].mode()[0], inplace=True)

# Fill missing fare in test set
test['Fare'].fillna(test['Fare'].median(), inplace=True)

```

## Step 4: Convert categorical text to numbers

```python
# Convert categorical to numeric
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})

train['Embarked'] = train['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
test['Embarked'] = test['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

## Step 5: Choose features and train the model

```python
# Choose features
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

X = train[features]
y = train['Survived']


```

## Step 6: Train a model

```python
# Import and train model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X, y)
```

## Step 7: Make predictions on the test data

```python
X_test = test[features]
predictions = model.predict(X_test)
```

## Step 8:Create and save submission file

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
```
